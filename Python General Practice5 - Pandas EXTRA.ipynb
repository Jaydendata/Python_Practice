    ## Regression

```python
from sklearn import linear_model
from sklearn import datasets
from sklearn.model_selection import train_test_split

import pandas as pd
import numpy as np
from pprint import pprint as pp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
housing_data = datasets.fetch_california_housing()
```

```python
pp(housing_data.DESCR)
```

```python
housing_data.data
```

```python
housing_data.feature_names
```

```python
housing_data.target
```

```python
df_housing = pd.DataFrame(data = housing_data.data,columns = housing_data.feature_names)
df_housing.head()
```

```python
df_housing['target'] = housing_data.target
df_housing.head()
```

```python
X = df_housing[['HouseAge']]
y = df_housing['target']

# x must be dataframe
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state= 42)

# random_state is to avoid different answers on the same machine each time run this program.
```

```python
lr = linear_model.LinearRegression()
lr.fit(X_train, y_train)
lr.score(X_test, y_test)

print(lr.coef_)
print(lr.intercept_)
```

```python
lr.predict(X_test)
```
